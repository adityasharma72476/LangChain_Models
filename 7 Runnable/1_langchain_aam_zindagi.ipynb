```python
import random
```

```python
class NakliLLM:

    def __init__(self):
        print('LLM created')

    def predict(self, prompt):

        response_list = [
            "New Delhi is the capital of India.",
            "IPL is Indian Premier League.",
            "Virat Kohli is the King of run chase."
        ]

        return {'response': random.choice(response_list)}
```

```python
class NakliPromptTemplate:

    def __init__(self, template, input_variable):
        self.template = template
        self.input_variable = input_variable

    def format(self, input_dict):
        return self.template.format(**input_dict)
```

```python
template = NakliPromptTemplate(
    template = "Write a {length} poem about {topic}",
    input_variable=['length','topic']
)
```

```python
class NakliLLM_chain:

    def __init__(self, llm, prompt):
        self.llm = llm
        self.prompt = prompt
    
    def run(self, input_dict):
        
        final_prompt = self.prompt.format(input_dict)
        result = self.llm.predict(final_prompt)

        return result['response']
```

```python
prompt = template.format({'length': "short",
                 'topic': 'Cricket'})
```

```python
llm = NakliLLM()
```

```python
chain = NakliLLM_chain(llm, template)
```

```python
chain.run({'length':'short','topic':'Cricket'})
```
